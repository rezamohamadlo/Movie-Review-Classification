```python
import tensorflow_datasets as tfds
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load the IMDB dataset
imdb, info = tfds.load('imdb_reviews', with_info=True, as_supervised=True)
train_data, test_data = imdb['train'], imdb['test']

# Initialize lists to hold training and testing sentences and labels
training_sentences = []
training_labels = []
testing_sentences = []
testing_labels = []

# Preprocess the training data
for s, l in train_data:
    training_sentences.append(s.numpy().decode('utf8'))  # Decode the bytes to string
    training_labels.append(l.numpy())  # Append the label

# Preprocess the testing data
for s, l in test_data:
    testing_sentences.append(s.numpy().decode('utf8'))  # Decode the bytes to string
    testing_labels.append(l.numpy())  # Append the label

# Convert labels to numpy arrays
training_labels_final = np.array(training_labels)
testing_labels_final = np.array(testing_labels)

# Define parameters for tokenization
vocab_size = 10000  # Limit the vocabulary size
max_length = 120    # Maximum length of sequences
trunc_type = 'post' # Truncate sequences at the end
oov_tok = "<OOV>"   # Out-of-vocabulary token

# Tokenize the training sentences
tokenizer = Tokenizer(num_words=vocab_size, oov_token=oov_tok)
tokenizer.fit_on_texts(training_sentences)  # Fit the tokenizer on training data
word_index = tokenizer.word_index  # Get the word index

# Convert training sentences to sequences and pad them
sequences = tokenizer.texts_to_sequences(training_sentences)
padded = pad_sequences(sequences, maxlen=max_length, truncating=trunc_type)

# Convert testing sentences to sequences and pad them
testing_sequences = tokenizer.texts_to_sequences(testing_sentences)
testing_padded = pad_sequences(testing_sequences, maxlen=max_length)
```
